```python
import pandas as pd
df = pd.read_csv("./script.csv")
```

```python
df.head()
```

```python
df["datetime"] = pd.to_datetime(df["datetime"])

df_slice = df[
    (df["datetime"].dt.month == 6) &
    (df["datetime"].dt.day == 15)
].set_index("datetime").between_time("18:30", "19:00")
```

```python
df_slice.head()
```

```python
import matplotlib.pyplot as plt

df_slice["requests"].plot()

plt.xlabel("datetime")
plt.ylabel("requests")
plt.title("Requests from 18:30 to 21:30 on June 15")
plt.show()
```
